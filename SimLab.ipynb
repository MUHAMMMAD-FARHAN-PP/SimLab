```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,1,2,3,4])
y=np.array([0,1,4,9,16])
a=np.array([0,1,2,3,4])
b=np.array([0,1,8,27,64])
plt.subplot(2,2,1)
plt.plot(x,y,color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(2,2,3)
plt.stem(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(2,2,2)
plt.plot(a,b,color='green')
plt.xlabel('a')
plt.ylabel('b')
plt.subplot(2,2,4)
plt.stem(a,b)
plt.xlabel('a')
plt.ylabel('b')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi,)
y = np.sin(x)
a = np.cos(x)
b = np.sin(2*x)
c = np.cos(2*x)

plt.figure(figsize=(10, 8))

plt.subplot(3,3,1)
plt.plot(x, y, color='blue')
plt.title('Sine Wave')
plt.xlabel('x (radians)')
plt.ylabel('sin(x)')

plt.subplot(3,3,3)
plt.plot(x, a, color='red')
plt.title('Cosine Wave')
plt.xlabel('x (radians)')
plt.ylabel('cos(x)')

plt.subplot(3,3,7)
plt.plot(x, b, color='green')
plt.xlabel('x (radians)')
plt.ylabel('sin(2x)')
plt.grid()

plt.subplot(3,3,9)
plt.plot(x, c, color='violet')
plt.xlabel('x (radians)')
plt.ylabel('cos(2x)')
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10)
a=2*np.exp(-0.2*x)

plt.subplot(1,1,1)
plt.plot(x,a)
plt.xlabel('x')
plt.ylabel('e^(-0.2x)')
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 4))

x = np.linspace(0, 10, 250)
y = np.sin(x)

noise = np.random.normal(0, 0.1, 250)
noisy_y = y + noise

plt.plot(x, noisy_y)
plt.xlabel('x')
plt.ylabel('Noisy Sine Wave')
plt.title('Noisy Sine Wave')
plt.grid()
plt.show()
```
